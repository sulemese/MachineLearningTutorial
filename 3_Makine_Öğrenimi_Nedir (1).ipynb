#MAKİNE ÖĞRENMESİ ALGORİTMALARI NASIL ÇALIŞIR

## 1. İleri Geçiş

İleri Geçişte, makine öğrenme algoritması giriş verilerini alır ve bir çıktı üretir. Model algoritmasına bağlı olarak tahminleri hesaplar.

## 2. Kayıp Fonksiyonu

Hata veya maliyet fonksiyonu olarak da bilinen kayıp fonksiyonu, modelin yaptığı tahminlerin doğruluğunu değerlendirmek için kullanılır. Fonksiyon, modelin tahmin edilen çıktısını gerçek çıktıyla karşılaştırır ve aralarındaki farkı hesaplar. Bu fark hata veya kayıp olarak bilinir. Modelin amacı, iç parametrelerini ayarlayarak hata veya kayıp fonksiyonunu en aza indirmektir.

## 3. Model Optimizasyon Süreci

 Model optimizasyon süreci, hata veya kayıp fonksiyonunu en aza indirecek şekilde modelin iç parametrelerini ayarlamanın yinelemeli sürecidir. Bu, gradyan iniş gibi bir optimizasyon algoritması kullanılarak yapılır . Optimizasyon algoritması, modelin parametrelerine göre hata fonksiyonunun eğimini hesaplar ve bu bilgiyi, hatayı azaltacak parametreleri ayarlamak için kullanır. Algoritma, hata tatmin edici bir düzeye indirilene kadar bu işlemi tekrarlar.


## 4. Test ve Değerlendirme

 Model, eğitim verileri üzerinde eğitildikten ve optimize edildikten sonra yeni, görünmeyen veriler üzerinde tahminlerde bulunmak için kullanılabilir. Modelin tahminlerinin doğruluğu, doğruluk, kesinlik, geri çağırma ve F1 puanı gibi çeşitli performans ölçümleri kullanılarak değerlendirilebilir.

#MAKİNE ÖĞRENİMİ PROJESİNİN YAŞAM DÖNGÜSÜ

1. Sorunları İnceleyin: İlk adım sorunu incelemektir. Bu adım, iş sorununu anlamayı ve modelin hedeflerini tanımlamayı içerir.

2. Veri Toplama: Sorun iyi tanımlandığında model için gerekli olan ilgili verileri toplayabiliriz. Veriler veritabanları, API'ler veya web kazıma gibi çeşitli kaynaklardan gelebilir.

3. Veri Hazırlama: Soruna ilişkin verilerimiz toplandığında. o zaman verileri doğru bir şekilde kontrol etmek ve model tarafından gizli kalıpları bulmak için kullanılabilmesi için istenen formatta hazırlamak iyi bir fikirdir. Bu, aşağıdaki adımlarda yapılabilir:
* Veri temizleme
* Veri Dönüşümü
* Açıklayıcı Veri Analizi ve Özellik Mühendisliği
* Eğitim ve test için veri kümesini bölün.

4. Model Seçimi: Bir sonraki adım problemimize uygun uygun makine öğrenme algoritmasını seçmektir. Bu adım, farklı algoritmaların güçlü ve zayıf yönlerinin bilinmesini gerektirir. Bazen birden fazla model kullanırız, sonuçlarını karşılaştırırız ve gereksinimlerimize göre en iyi modeli seçeriz.

Derin öğrenme durumunda, giriş ve çıkış boyutu, her katmandaki düğüm sayısı, kayıp fonksiyonu, gradyan iniş optimizasyonu vb. ile birlikte katman bazında mimariyi tanımlamamız gerekir.
Daha sonra model önceden işlenmiş veri seti kullanılarak eğitilir.

7. Model Ayarlaması: Değerlendirme sonuçlarına göre, performansını artırmak için modelin ayarlanması veya optimize edilmesi gerekebilir. Bu, modelin hiperparametrelerinin ayarlanmasını içerir.

8. Dağıtım: Model eğitildikten ve ayarlandıktan sonra, yeni verilerle ilgili tahminlerde bulunmak için üretim ortamında dağıtılabilir. Bu adım, modelin mevcut bir yazılım sistemine entegre edilmesini veya model için yeni bir sistem oluşturulmasını gerektirir.

9. İzleme ve Bakım: Son olarak, modelin üretim ortamındaki performansını izlemek ve gerektiğinde bakım görevlerini gerçekleştirmek önemlidir. Bu, veri kaymasının izlenmesini, gerektiğinde modelin yeniden eğitilmesini ve yeni veriler elde edildikçe modelin güncellenmesini içerir.

# MAKİNE ÖĞRENMESİ TÜRLERİ

## 1. Denetimli Makine Öğrenimi

**Denetimli öğrenme, algoritmanın etiketli veri kümesi üzerinde eğitildiği bir makine öğrenimi türüdür.** Etiketli eğitim verilerine dayanarak giriş özelliklerini hedeflerle eşlemeyi öğrenir. Denetimli öğrenmede, algoritmaya giriş özellikleri ve karşılık gelen çıkış etiketleri sağlanır ve**yeni, görünmeyen veriler üzerinde tahminler yapmak için bu verilerden genelleme yapmayı öğrenir.**

Denetimli öğrenmenin iki ana türü vardır:

1. **Regresyon : Regresyon, algoritmanın giriş özelliklerine dayalı olarak sürekli değerleri tahmin etmeyi öğrendiği bir denetimli öğrenme türüdür.** Regresyondaki çıktı etiketleri hisse senedi fiyatları ve konut fiyatları gibi sürekli değerlerdir. Makine öğrenimindeki farklı regresyon algoritmaları şunlardır: Doğrusal Regresyon, Polinom Regresyon, Ridge Regresyon, Karar Ağacı Regresyonu, Rastgele Orman Regresyonu, Destek Vektör Regresyonu vb.

2. **Sınıflandırma : Sınıflandırma, algoritmanın giriş verilerini, giriş özelliklerine göre belirli bir kategoriye veya sınıfa atamayı öğrendiği bir denetimli öğrenme türüdür.** Sınıflandırmadaki çıkış etiketleri ayrık değerlerdir. Sınıflandırma algoritmaları, çıktının iki olası sınıftan biri olduğu ikili veya çıktının birkaç sınıftan biri olduğu çoklu sınıf olabilir. Makine öğrenimindeki farklı Sınıflandırma algoritmaları şunlardır: Lojistik Regresyon, Naif Bayes, Karar Ağacı, Destek Vektör Makinesi (SVM), K-En Yakın Komşular (KNN), vb.

## 2. Denetimsiz Makine Öğrenimi


Denetimsiz öğrenme, algoritmanın, etiketli örnekler kullanılmadan  açıkça eğitilmeden verilerdeki kalıpları tanımayı öğrendiği bir makine öğrenimi türüdür. Denetimsiz öğrenmenin amacı, verilerdeki temel yapıyı veya dağılımı keşfetmektir.

Denetimsiz öğrenmenin iki ana türü vardır:


1. **Kümeleme : Kümeleme algoritmaları benzer veri noktalarını özelliklerine göre gruplandırır.** Amaç, birbirine benzer ancak diğer gruplardan farklı olan veri noktalarından oluşan grupları veya kümeleri tanımlamaktır. Bazı popüler kümeleme algoritmaları arasında K-araçları, Hiyerarşik kümeleme ve DBSCAN bulunur.


2. **Boyut azaltma: Boyut azaltma algoritmaları, orijinal bilgilerin mümkün olduğu kadar çoğunu korurken bir veri kümesindeki giriş değişkenlerinin sayısını azaltır.** Bu, bir veri kümesinin karmaşıklığını azaltmak ve görselleştirmeyi ve analiz etmeyi kolaylaştırmak için kullanışlıdır. Bazı popüler boyut azaltma algoritmaları arasında Temel Bileşen Analizi (PCA), t-SNE ve Otomatik Kodlayıcılar bulunur.

## 3. Takviyeli Makine Öğrenimi

**Takviyeli öğrenme, bir aracının eylemler gerçekleştirerek ve eylemlerine bağlı olarak ödüller veya cezalar alarak bir ortamla etkileşime girmeyi öğrendiği bir tür makine öğrenimidir**. Takviyeli öğrenmenin amacı, zaman içinde beklenen kümülatif ödülü en üst düzeye çıkaran, durumlardan eylemlere haritalama olan bir politikayı öğrenmektir.

Takviyeli öğrenmenin iki ana türü vardır:


1. **Model tabanlı takviyeli öğrenme: Model bazlı takviyeli öğrenmede, aracı, durumlar arasındaki geçiş olasılıkları ve her durum-eylem çiftiyle ilişkili ödüller dahil olmak üzere ortamın bir modelini öğrenir**. Temsilci daha sonra beklenen ödülü en üst düzeye çıkarmak amacıyla eylemlerini planlamak için bu modeli kullanır. Bazı popüler model tabanlı takviyeli öğrenme algoritmaları arasında Değer Yinelemesi ve Politika Yinelemesi bulunur.


2. **Modelsiz takviyeli öğrenme : Modelsiz takviyeli öğrenmede, aracı, açıkça bir ortam modeli oluşturmadan, doğrudan deneyimden bir politika öğrenir.** Temsilci çevreyle etkileşime girer ve aldığı ödüllere göre politikasını günceller. Bazı popüler modelsiz takviyeli öğrenme algoritmaları arasında Q-Learning, SARSA ve Deep Reinforcement Learning bulunur.

#MAKİNE ÖĞRENİMİNE DUYULAN İHTİYAÇ

Makine öğrenimi önemlidir çünkü bilgisayarların verilerden öğrenmesine ve açıkça programlanmadan belirli görevlerdeki performanslarını artırmasına olanak tanır. Verilerden öğrenme ve yeni durumlara uyum sağlama yeteneği, makine öğrenimini özellikle büyük miktarda veri, karmaşık karar verme ve dinamik ortamlar içeren görevler için yararlı kılar.

Makine öğreniminin kullanıldığı bazı spesifik alanlar şunlardır:

* Tahmine dayalı modelleme: Makine öğrenimi, işletmelerin daha iyi kararlar almasına yardımcı olabilecek tahmine dayalı modeller oluşturmak için kullanılabilir. Örneğin, makine öğrenimi, hangi müşterilerin belirli bir ürünü satın alma olasılığının yüksek olduğunu veya hangi hastaların belirli bir hastalığa yakalanma olasılığının yüksek olduğunu tahmin etmek için kullanılabilir.


* Doğal dil işleme: Makine öğrenimi, insan dilini anlayabilen ve yorumlayabilen sistemler oluşturmak için kullanılır. Bu, ses tanıma, sohbet robotları ve dil çevirisi gibi uygulamalar için önemlidir.


* Bilgisayarla görme: Makine öğrenimi, görüntüleri ve videoları tanıyabilen ve yorumlayabilen sistemler oluşturmak için kullanılır. Bu, sürücüsüz arabalar, gözetim sistemleri ve tıbbi görüntüleme gibi uygulamalar için önemlidir.

* Dolandırıcılık tespiti: Makine öğrenimi, finansal işlemlerde, çevrimiçi reklamlarda ve diğer alanlardaki dolandırıcılık davranışlarını tespit etmek için kullanılabilir.


* Öneri sistemleri: Makine öğrenimi, kullanıcılara geçmiş davranışlarına ve tercihlerine göre ürün, hizmet veya içerik öneren öneri sistemleri oluşturmak için kullanılabilir.

Genel olarak makine öğrenimi, verileri daha iyi kullanmalarına, karar verme süreçlerini iyileştirmelerine ve müşterilerine daha kişiselleştirilmiş deneyimler sunmalarına olanak tanıdığı için birçok işletme ve sektör için önemli bir araç haline geldi.
