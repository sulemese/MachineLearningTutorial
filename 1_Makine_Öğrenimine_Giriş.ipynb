#MAKİNE ÖĞRENİMİ

image.png

**"Makine Öğrenimi" : bilgisayarlara açıkça programlanmadan öğrenme yeteneği kazandıran yapay zekanın alt dalıdır.**


**"Makine Öğrenimi" : Örnek veriler ve geçmiş deneyimlerin kullanılarak  performansı optimize etmek için bilgisayarların programlanmasıdır.**

**"Makine Öğrenimi" : bilgisayar sistemlerinin veriyle beslenerek örüntüleri tanıması, öğrenmesi ve kararlar alması için kullanılan bir yapay zeka alanıdır.**

**"Evrensel bir tanım olarak, makine öğrenimi" :  bir algoritmanın veri setleri üzerindeki istatistiksel örüntüleri algılaması ve bu örüntülerden genelleme yaparak gelecekteki verileri tahmin etmesi sürecidir.**

Makine öğrenimi genellikle bilgisayar sistemlerinin veriyle beslenerek örüntüleri tanıması, öğrenmesi ve kararlar alması için kullanılan bir yapay zeka alanıdır. Evrensel bir tanım olarak, makine öğrenimi, bir algoritmanın veri setleri üzerindeki istatistiksel örüntüleri algılaması ve bu örüntülerden genelleme yaparak gelecekteki verileri tahmin etmesi sürecidir. Bu tanım altında, makine öğrenimi çeşitli teknikler, algoritmalar ve modeller kullanarak karmaşık problemleri çözmek için bir araç olarak kullanılabilir. Bu, sınıflandırma, regresyon, kümeleme, boyut azaltma gibi birçok farklı görevi içerebilir.

Bilgisayar oyunları ve yapay zeka alanında ilk Amerikalı liderlerden biri olan Arthur Samuel, 1959'da IBM'deyken "Makine Öğrenimi" terimini icat etti. Makine öğrenimini “bilgisayarlara açıkça programlanmadan öğrenme yeteneği kazandıran çalışma alanı” olarak tanımladı. Ancak makine öğreniminin evrensel olarak kabul edilmiş bir tanımı yoktur. Farklı yazarlar terimi farklı şekilde tanımlarlar. Aşağıda iki tanım daha veriyoruz.

Makine öğrenimi, örnek verileri veya geçmiş deneyimleri kullanarak bir performans kriterini optimize etmek için bilgisayarların programlanmasıdır. Bazı parametrelere kadar tanımlanmış bir modelimiz var ve öğrenme, eğitim verilerini veya geçmiş deneyimleri kullanarak modelin parametrelerini optimize etmek için bir bilgisayar programının çalıştırılmasıdır. Model, gelecekte tahminlerde bulunmak için öngörücü olabilir veya verilerden bilgi elde etmek için tanımlayıcı olabilir.

Makine öğrenmesi olarak bilinen çalışma alanı, deneyimle otomatik olarak gelişen bilgisayar programlarının nasıl oluşturulacağı sorusuyla ilgilidir.

Makine öğrenimi, bilgisayarların görevlerdeki performanslarını deneyim yoluyla artırmasını sağlayan algoritmaların ve istatistiksel modellerin geliştirilmesini içeren, yapay zekanın bir alt alanıdır. Bu algoritmalar ve modeller, verilerden öğrenmek ve açık talimatlar olmaksızın tahminlerde bulunmak veya kararlar almak üzere tasarlanmıştır. Denetimli öğrenme, denetimsiz öğrenme ve takviyeli öğrenme dahil olmak üzere çeşitli makine öğrenimi türleri vardır. Denetimli öğrenme, bir modelin etiketli veriler üzerinde eğitilmesini içerirken, denetimsiz öğrenme, bir modelin etiketlenmemiş veriler üzerinde eğitilmesini içerir. Takviyeli öğrenme, bir modelin deneme yanılma yoluyla eğitilmesini içerir. Makine öğrenimi, görüntü ve konuşma tanıma, doğal dil işleme ve öneri sistemleri dahil olmak üzere çok çeşitli uygulamalarda kullanılır.

#ÖĞRENME

"Öğrenme " : Bir bilgisayar programının , P ile ölçülen T görevlerindeki performansı, E deneyimiyle birlikte artıyorsa, bazı T görev sınıflarına ve P performans ölçümüne ilişkin deneyim E'den öğrendiği  söylenir .

**"Öğrenme" : bir algoritmanın veri setleri üzerinde istatistiksel örüntüleri tanıması ve bu örüntülerden genelleme yaparak gelecekteki verileri tahmin etmesi sürecidir.**

"Öğrenme" : algoritmanın deneyimlerle eğitilmesi ve bu deneyimlerden bilgi edinerek gelecekteki kararlarını ve davranışlarını iyileştirmesiyle gerçekleşir.

**"Öğrenme Programı" : Deneyimlerden öğrenen bir bilgisayar programına "makine öğrenme programı" veya kısaca "öğrenme programı" denir.**

**Örnek**

El yazısı tanıma öğrenme problemi

Görev T: Resimlerdeki el yazısıyla yazılan kelimeleri tanıma ve sınıflandırma

Performans P : Doğru sınıflandırılan kelimelerin yüzdesi

Eğitim deneyimi E : Verilen sınıflandırmalara sahip el yazısıyla yazılmış kelimelerden oluşan bir veri seti



**Örnek**

Bir robot sürüş öğrenme problemi

Görev T: Görüş sensörlerini kullanarak otoyolda sürüş yapmak

Performans P : Hatadan önce kat edilen ortalama mesafe

Eğitim deneyimi E: Bir insan sürücüyü gözlemlerken kaydedilen bir dizi görüntü ve direksiyon komutu

#MAKİNE ÖĞRENİMİ SINIFLANDIRILMASI

Makine öğrenimi uygulamaları, bir öğrenme sistemine sunulan öğrenme "sinyali" veya "yanıtı"nın niteliğine bağlı olarak dört ana kategoride sınıflandırılır:

##Denetimli Öğrenme

**"Denetimli öğrenme" : örnek girdi-çıktı çiftlerine dayalı olarak bir girdiyi bir çıktıyla eşleyen bir fonksiyonun öğrenildiği makine öğrenimi görevidir.** Verilen veriler etiketlenir. Hem sınıflandırma hem de regresyon problemleri denetimli öğrenme problemleridir.

**Örnek**

Bir kliniğe giren hastalarla ilgili aşağıdaki verileri göz önünde bulundurun. Veriler hastaların cinsiyet ve yaşlarından oluşuyor ve her hasta “sağlıklı” veya “hasta” olarak etiketleniyor.

##Denetimsiz Öğrenme

**"Denetimsiz öğrenme" : etiketlenmiş yanıtlar olmadan giriş verilerinden oluşan veri kümelerinden çıkarımlar yapmak için kullanılan bir tür makine öğrenimi algoritmasıdır.** Denetimsiz öğrenme algoritmalarında gözlemlere sınıflandırma veya kategorizasyon dahil edilmez.

**Örnek**
Bir kliniğe giren hastalarla ilgili aşağıdaki verileri düşünün. Veriler hastaların cinsiyet ve yaşlarından oluşmaktadır.

Bir tür öğrenme olarak, insanların belirli nesne veya olayların aynı sınıftan olduğunu anlamak için kullandığı yöntemlere benzer; örneğin nesneler arasındaki benzerlik derecesini gözlemlemek gibi. Web'de pazarlama otomasyonu biçiminde bulduğunuz bazı öneri sistemleri bu tür öğrenmeye dayanmaktadır.

##Takviyeli Öğrenme

**"Takviyeli öğrenme" : bir aracının dünyada ödüllerini en üst düzeye çıkaracak şekilde hareket etmesini sağlama sorunudur.**

Bir öğrenciye çoğu makine öğrenimi biçiminde olduğu gibi hangi eylemleri gerçekleştirmesi gerektiği söylenmez, bunun yerine hangi eylemlerin en fazla ödülü getirdiğini deneyerek keşfetmesi gerekir. Örneğin — Bir köpeğe yeni bir numara öğretmeyi düşünün: Ona ne yapması gerektiğini, ne yapmaması gerektiğini söyleyemeyiz, ancak doğru/yanlış şeyi yaparsa onu ödüllendirebilir/cezalandırabiliriz.

 Videoyu izlerken programın başlangıçta ne kadar beceriksiz ve beceriksiz olduğunu, ancak şampiyon olana kadar eğitimle istikrarlı bir şekilde geliştiğini fark edeceksiniz.

##Yarı Denetimli Öğrenme

Tamamlanmamış bir eğitim sinyalinin verildiği yer: hedef çıktıların bir kısmının (genellikle çoğunun) eksik olduğu bir eğitim seti. Bu prensibin, hedeflerin bir kısmının eksik olması dışında, öğrenme zamanında problem örneklerinin tamamının bilindiği, Transdüksiyon olarak bilinen özel bir durumu vardır. **Yarı denetimli öğrenme, eğitim sırasında küçük etiketli verileri büyük miktarda etiketlenmemiş verilerle birleştiren bir makine öğrenimi yaklaşımıdır.** Yarı denetimli öğrenme, denetimsiz öğrenme ile denetimli öğrenme arasında yer alır.

#GEREKLİ ÇIKTIYA GÖRE MAKİNE ÖĞRENİMİ SINIFLANDIRILMASI

Makine öğrenimi görevlerinin başka bir sınıflandırması, makine öğrenimi sisteminin arzu edilen çıktısı dikkate alındığında ortaya çıkar:  

##Sınıflandırma

**Girdiler iki veya daha fazla sınıfa bölündüğünde, öğrenci, görünmeyen girdileri bu sınıfların bir veya daha fazlasına (çok etiketli sınıflandırma) atayan bir model üretmelidir.** Bu genellikle denetimli bir şekilde ele alınır. Spam filtreleme, girişlerin e-posta (veya diğer) mesajları olduğu ve sınıfların "spam" ve "spam değil" olduğu bir sınıflandırma örneğidir.

##Regresyon

Denetimli öğrenme problemidir.  Çıktılar ayrık değil süreklidir.

##Kümeleme

Bir dizi girdinin gruplara bölünmesi gerektiği zaman kullanılır.  Sınıflandırmanın aksine, gruplar önceden bilinmediğinden bu genellikle denetimsiz öğrenme problemi haline gelir.
