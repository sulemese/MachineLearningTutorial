#MAKİNE ÖĞRENİMİ TÜRLERİ

Makine öğrenimi , bilgisayarların verilerden öğrenmesine ve her görev için açıkça programlanmadan önceki deneyimlerden yararlanmasına olanak tanıyan modeller ve algoritmalar geliştirmeye odaklanan Yapay Zekanın bir dalıdır . Basit bir ifadeyle ML, verilerden öğrenerek sistemlere insanlar gibi düşünmeyi ve anlamayı öğretir.

Bu makalede gelecekteki gereksinimler için önemli olan çeşitli makine öğrenimi algoritmalarını inceleyeceğiz . Makine öğrenimi genel olarak geçmiş deneyimlerden ders almaya ve zaman içinde performansı artırmaya yönelik bir eğitim sistemidir. Makine öğrenimi, büyük miktarda verinin tahmin edilmesine yardımcı olur. Kârlı fırsatlar elde etmek için hızlı ve doğru sonuçlar sunmaya yardımcı olur.

Her biri özel özelliklere ve uygulamalara sahip çeşitli makine öğrenimi türleri vardır. Makine öğrenimi algoritmalarının ana türlerinden bazıları şunlardır:

1. Denetimli Makine Öğrenimi
2. Denetimsiz Makine Öğrenimi
3. Yarı Denetimli Makine Öğrenimi
4. Takviyeli Öğrenme

##1. Denetimli Makine Öğrenimi


image.png

Denetimli öğrenme, bir modelin "Etiketli Veri Kümesi" üzerinde eğitilmesi olarak tanımlanır . Etiketli veri kümeleri hem giriş hem de çıkış parametrelerine sahiptir. Denetimli Öğrenme algoritmaları, girdiler ve doğru çıktılar arasındaki noktaları haritalamayı öğrenir. Etiketlenmiş hem eğitim hem de doğrulama veri kümelerine sahiptir.

**Örnek**

Kediler ve köpekler arasında ayrım yapmak için bir görüntü sınıflandırıcı oluşturmanız gereken bir senaryo düşünün. Köpek ve kedi etiketli görsellerin veri setlerini algoritmaya beslerseniz, makine bu etiketli görsellerden köpek veya kedi arasında sınıflandırma yapmayı öğrenecektir. Daha önce hiç görmediği yeni köpek veya kedi görsellerini girdiğimizde öğrenilen algoritmaları kullanacak ve köpek mi kedi mi olduğunu tahmin edecek. Denetimli öğrenme bu şekilde çalışır ve bu özellikle bir görüntü sınıflandırmasıdır.

Aşağıda belirtilen denetimli öğrenmenin iki ana kategorisi vardır:

1. Sınıflandırma
2. Regresyon

**Sınıflandırma**

Sınıflandırma , ayrı sınıfları veya etiketleri temsil eden kategorik hedef değişkenlerin tahmin edilmesiyle ilgilidir. Örneğin, e-postaları spam veya spam değil olarak sınıflandırmak veya bir hastanın yüksek kalp hastalığı riskine sahip olup olmadığını tahmin etmek. Sınıflandırma algoritmaları, giriş özelliklerini önceden tanımlanmış sınıflardan birine eşlemeyi öğrenir.

İşte bazı sınıflandırma algoritmaları:

* Logistic Regression
* Support Vector Machine
* Random Forest
* Decision Tree
* K-Nearest Neighbors (KNN)
* Naive Bayes

**Regresyon**

Regresyon ise sayısal değerleri temsil eden sürekli hedef değişkenlerin tahmin edilmesiyle ilgilenir. Örneğin, bir evin fiyatını büyüklüğüne, konumuna ve olanaklarına göre tahmin etmek veya bir ürünün satışını tahmin etmek. Regresyon algoritmaları, giriş özelliklerini sürekli bir sayısal değere eşlemeyi öğrenir.

İşte bazı regresyon algoritmaları:

* Linear Regression
* Polynomial Regression
* Ridge Regression
* Lasso Regression
* Decision tree
* Random Forest

**Denetimli Makine Öğreniminin Avantajları**

1. Denetimli Öğrenme modelleri, etiketli veriler üzerinde eğitildikleri için yüksek doğruluğa sahip olabilir .

2. Denetimli öğrenme modellerinde karar verme süreci genellikle yorumlanabilir.

3. Sıfırdan yeni modeller geliştirirken zamandan ve kaynaklardan tasarruf sağlayan önceden eğitilmiş modellerde sıklıkla kullanılabilir.

**Denetimli Makine Öğreniminin Dezavantajları**

1. Kalıpları bilme konusunda sınırlamaları vardır ve eğitim verilerinde bulunmayan, görülmeyen veya beklenmeyen kalıplarla mücadele edebilir.

2. Yalnızca etiketli verilere dayandığı için zaman alıcı ve maliyetli olabilir .
Yeni verilere dayanarak zayıf genellemelere yol açabilir.

**Denetimli Öğrenme Uygulamaları**


Denetimli öğrenme, aşağıdakiler de dahil olmak üzere çok çeşitli uygulamalarda kullanılır:


* Görüntü sınıflandırması : Görüntülerdeki nesneleri, yüzleri ve diğer
* Doğal dil işleme: Metinden duygu, varlık ve ilişkiler gibi bilgileri çıkarın.
* Konuşma tanıma : Konuşma dilini metne dönüştürün.
* Öneri sistemleri : Kullanıcılara kişiselleştirilmiş önerilerde bulunun.
* Tahmine dayalı analitik : Satışlar, müşteri kaybı ve hisse senedi fiyatları gibi sonuçları tahmin edin.
* Tıbbi teşhis : Hastalıkları ve diğer tıbbi durumları tespit edin.
* Dolandırıcılık tespiti : Dolandırıcılık işlemlerini tanımlayın.
* Otonom araçlar : Ortamdaki nesneleri tanır ve onlara yanıt verir.
* E-posta spam tespiti : E-postaları spam veya spam değil olarak sınıflandırın.
* Üretimde kalite kontrolü : Ürünleri kusurlara karşı inceleyin.
* Kredi puanlaması : Borçlunun krediyi temerrüde düşürme riskini değerlendirin.
* Oyun : Karakterleri tanıyın, oyuncu davranışını analiz edin ve NPC'ler oluşturun.
* Müşteri desteği : Müşteri destek görevlerini otomatikleştirin.
* Hava tahmini : Sıcaklık, yağış ve diğer meteorolojik parametreler için tahminler yapın.
* Spor analitiği : Oyuncu performansını analiz edin, oyun tahminleri yapın ve stratejileri optimize edin.



## Denetimsiz Makine Öğrenimi

Denetimsiz Öğrenme Denetimsiz öğrenme, bir algoritmanın etiketlenmemiş verileri kullanarak kalıpları ve ilişkileri keşfettiği bir tür makine öğrenme tekniğidir. Denetimli öğrenmenin aksine, denetimsiz öğrenme algoritmaya etiketli hedef çıktılar sağlamayı gerektirmez. **Denetimsiz öğrenmenin temel amacı genellikle veriler içindeki gizli kalıpları, benzerlikleri veya kümeleri keşfetmektir;** bunlar daha sonra veri araştırması, görselleştirme, boyut azaltma ve daha fazlası gibi çeşitli amaçlar için kullanılabilir.

image.png

**Örnek**

Mağazadan yaptığınız alışverişlere ilişkin bilgileri içeren bir veri kümeniz olduğunu varsayalım. Algoritma, kümeleme yoluyla siz ve diğer müşteriler arasında aynı satın alma davranışını gruplandırabilir ve bu da önceden tanımlanmış etiketler olmadan potansiyel müşterileri ortaya çıkarabilir. Bu tür bilgiler, işletmelerin hedef müşterilere ulaşmasına ve aykırı değerleri belirlemesine yardımcı olabilir.

2 ana kategoriden meydana gelir

1. Kümeleme (Clustering)
2. Birliktelik (Association)

**Kümeleme**


Kümeleme, veri noktalarını benzerliklerine göre kümeler halinde gruplama işlemidir. Bu teknik, etiketli örneklere gerek kalmadan verilerdeki kalıpları ve ilişkileri tanımlamak için kullanışlıdır.

İşte bazı kümeleme algoritmaları:
* K-Means Clustering algorithm
* Mean-shift algorithm
* DBSCAN Algorithm
* Principal Component Analysis
* Independent Component Analysis


**Birliktelik (Association)**


Birliktelik kuralı öğrenme , bir veri kümesindeki öğeler arasındaki ilişkileri keşfetmeye yönelik bir tekniktir. Bir öğenin varlığını belirten kuralları belirli bir olasılıkla başka bir öğenin varlığını ima eder.

Bazı birliktelik kuralı öğrenme algoritmaları şunlardır:
* Apriori Algorithm
* Eclat
* FP-growth Algorithm


**Denetimsiz Makine Öğreniminin Avantajları**
1. Veriler arasındaki gizli kalıpları ve çeşitli ilişkileri keşfetmeye yardımcı olur.

2. Müşteri segmentasyonu, anormallik tespiti ve veri araştırması gibi görevler için kullanılır .

3. Etiketlenmiş verilere ihtiyaç duymaz ve veri etiketleme zahmetini azaltır.

**Denetimsiz Makine Öğreniminin Dezavantajları**

1. Etiketleri kullanmadan model çıktısının kalitesini tahmin etmek zor olabilir.
2. Küme Yorumlanabilirliği açık olmayabilir ve anlamlı yorumlara sahip olmayabilir.
3. Ham verilerden anlamlı özellikler çıkarmak için kullanılabilecek otomatik kodlayıcılar ve boyut azaltma gibi tekniklere sahiptir .

**Denetimsiz Öğrenme Uygulamaları**

Denetimsiz öğrenmenin bazı yaygın uygulamaları şunlardır:

* Kümeleme : Benzer veri noktalarını kümeler halinde gruplayın.
* Anormallik tespiti : Verilerdeki aykırı değerleri veya anormallikleri belirleyin.
* Boyut azaltma : Temel bilgileri korurken verinin boyutunu azaltın.
* Öneri sistemleri : Kullanıcılara geçmiş davranışlarına veya tercihlerine göre ürünler, filmler veya içerik önerin.
* Konu modelleme : Bir belge koleksiyonundaki gizli konuları keşfedin.
* Yoğunluk tahmini : Verilerin olasılık yoğunluk fonksiyonunu tahmin edin.
* Resim ve video sıkıştırma : Multimedya içeriği için gereken depolama miktarını azaltın.
* Veri ön işleme : Veri temizleme, eksik değerlerin atanması ve veri ölçeklendirme gibi veri ön işleme görevlerine yardımcı olur.
* Pazar sepeti analizi : Ürünler arasındaki ilişkileri keşfedin.
* Genomik veri analizi : Benzer ifade profillerine sahip kalıpları veya grup genlerini belirleyin.
* Görüntü segmentasyonu : Görüntüleri anlamlı bölgelere ayırın.
* Sosyal ağlarda topluluk tespiti : Benzer ilgi alanlarına veya bağlantılara sahip toplulukları veya birey gruplarını tanımlayın.
* Müşteri davranışı analizi : Daha iyi pazarlama ve ürün önerileri için kalıpları ve öngörüleri ortaya çıkarın.
* İçerik önerisi : Kullanıcılara benzer öğelerin önerilmesini kolaylaştırmak için içeriği sınıflandırın ve etiketleyin.
* Keşif amaçlı veri analizi (EDA) : Belirli görevleri tanımlamadan önce verileri keşfedin ve öngörüler elde edin.

## Yarı Denetimli Makine Öğrenmesi

Yarı Denetimli öğrenme, denetimli ve denetimsiz öğrenme arasında çalışan, böylece hem etiketli hem de etiketsiz verileri kullanan bir makine öğrenme algoritmasıdır . Etiketli verileri elde etmenin maliyetli, zaman alıcı veya kaynak yoğun olduğu durumlarda özellikle kullanışlıdır. Bu yaklaşım, veri kümesinin pahalı ve zaman alıcı olduğu durumlarda faydalıdır. Yarı denetimli öğrenme, etiketli verilerin eğitilmesi veya ondan öğrenilmesi için beceriler ve ilgili kaynaklar gerektirdiğinde seçilir.

Bu teknikleri, bir kısmı etiketlenmiş ve geri kalan büyük kısmı etiketlenmemiş verilerle uğraşırken kullanırız. Etiketleri tahmin etmek için denetimsiz teknikleri kullanabilir ve ardından bu etiketleri denetimli tekniklere besleyebiliriz. Bu teknik çoğunlukla, genellikle tüm görüntülerin etiketlenmediği görüntü veri kümeleri durumunda uygulanabilir.

image.png

**Örnek**

Bir dil çeviri modeli oluşturduğumuzu düşünün; her cümle çifti için çevirilerin etiketlenmesi kaynak açısından yoğun olabilir. Modellerin etiketli ve etiketsiz cümle çiftlerinden öğrenmesine olanak tanıyarak onları daha doğru hale getirir. Bu teknik, makine çevirisi hizmetlerinin kalitesinde önemli gelişmelere yol açmıştır.

**Yarı Denetimli Öğrenme Yöntemi Türleri**


1.Graph-based semi-supervised learning(Grafik tabanlı yarı denetimli öğrenme)

 Bu yaklaşım, veri noktaları arasındaki ilişkileri temsil etmek için bir grafik kullanır. Grafik daha sonra etiketleri etiketli veri noktalarından etiketlenmemiş veri noktalarına yaymak için kullanılır.


2. Label propagation(Etiket yayılımı)

 Bu yaklaşım, veri noktaları arasındaki benzerliklere dayalı olarak etiketleri etiketli veri noktalarından etiketlenmemiş veri noktalarına yinelemeli olarak yayar.

3. Co-training(Ortak eğitim)

 Bu yaklaşım, etiketlenmemiş verilerin farklı alt kümeleri üzerinde iki farklı makine öğrenimi modelini eğitir. Daha sonra iki model birbirlerinin tahminlerini etiketlemek için kullanılır.

4. Self-training(Kendi kendine eğitim)

 Bu yaklaşım, etiketli veriler üzerinde bir makine öğrenimi modelini eğitir ve ardından modeli, etiketlenmemiş veriler için etiketleri tahmin etmek için kullanır. Model daha sonra etiketli veriler ve etiketlenmemiş veriler için tahmin edilen etiketler üzerinde yeniden eğitilir.

5. Generative adversarial networks (GANs/Üretken çekişmeli ağlar)

 GAN'lar, sentetik veriler oluşturmak için kullanılabilecek bir tür derin öğrenme algoritmasıdır. GAN'lar, iki sinir ağını, bir üreteci ve bir ayırıcıyı eğiterek yarı denetimli öğrenme için etiketlenmemiş veriler oluşturmak için kullanılabilir.

**Yarı Denetimli Makine Öğreniminin Avantajları**

1. Hem etiketli hem de etiketsiz verileri aldığından denetimli öğrenmeye kıyasla daha iyi genellemeye yol açar .

2. Çok çeşitli verilere uygulanabilir.

**Yarı Denetimli Makine Öğreniminin Dezavantajları**

1. Yarı denetimli yöntemlerin uygulanması diğer yaklaşımlara göre daha karmaşık olabilir.
2. Her zaman mevcut olmayabilecek veya elde edilmesi kolay olmayabilecek bazı etiketli verilere hâlâ ihtiyaç duyulmaktadır .

3. Etiketlenmemiş veriler model performansını buna göre etkileyebilir.

**Yarı Denetimli Öğrenme Uygulamaları**

Yarı denetimli öğrenmenin bazı yaygın uygulamaları şunlardır:

1. Görüntü Sınıflandırma ve Nesne Tanıma : Küçük bir etiketli görüntü kümesini daha büyük bir etiketsiz görüntü kümesiyle birleştirerek modellerin doğruluğunu artırın.

2. Doğal Dil İşleme (NLP) : Küçük bir etiketli metin verisi kümesini çok miktarda etiketlenmemiş metinle birleştirerek dil modellerinin ve sınıflandırıcıların performansını artırın.

3. Konuşma Tanıma: Sınırlı miktarda kopyalanmış konuşma verisinden ve daha kapsamlı bir etiketlenmemiş ses kümesinden yararlanarak konuşma tanımanın doğruluğunu artırın.

4. Öneri Sistemleri : Az sayıdaki kullanıcı öğesi etkileşimlerini (etiketli veriler) çok sayıda etiketlenmemiş kullanıcı davranışı verileriyle tamamlayarak kişiselleştirilmiş önerilerin doğruluğunu artırın.

5. Sağlık Hizmetleri ve Tıbbi Görüntüleme : Daha büyük bir etiketsiz görüntü kümesinin yanı sıra küçük bir etiketli tıbbi görüntü kümesi kullanarak tıbbi görüntü analizini geliştirin.

## 4. Takviyeli Makine Öğrenimi


**Eylemler üreterek ve hataları keşfederek çevreyle etkileşime giren bir öğrenme yöntemidir.** Deneme, yanılma ve gecikme, takviyeli öğrenmenin en alakalı özellikleridir. Bu teknikte model, davranışı veya modeli öğrenmek için Ödül Geri Bildirimini kullanarak performansını artırmaya devam eder. Bu algoritmalar belirli bir soruna özeldir; örneğin, Go Oyununda giderek daha iyi performans gösteren bir botun insanlarla ve hatta kendisiyle yarıştığı Google Kendi Kendine Sürüş arabası, AlphaGo gibi. Verileri her beslediğimizde, onlar öğrenir ve verileri kendi bilgilerine, yani eğitim verisine eklerler. Yani ne kadar çok öğrenirse o kadar iyi eğitilir ve dolayısıyla deneyimlenir.

**takviyeli öğrenme algoritmalarından bazıları**

1. Q-öğrenme: Q-öğrenme, durumları eylemlerle eşleyen bir Q işlevini öğrenen, modelsiz bir RL algoritmasıdır. Q işlevi, belirli bir durumda belirli bir eylemi gerçekleştirmenin beklenen ödülünü tahmin eder.

2. SARSA (Durum-Eylem-Ödül-Durum-Eylem): SARSA, bir Q fonksiyonunu öğrenen, modelden bağımsız başka bir RL algoritmasıdır. Bununla birlikte, Q-öğrenmesinden farklı olarak SARSA, Q-fonksiyonunu optimal eylem yerine gerçekte gerçekleştirilen eyleme göre günceller.

3. Derin Q-öğrenme : Derin Q-öğrenme, Q-öğrenme ve derin öğrenmenin birleşimidir. Derin Q-öğrenme, Q fonksiyonunu temsil etmek için bir sinir ağı kullanır ve bu da onun durumlar ve eylemler arasındaki karmaşık ilişkileri öğrenmesine olanak tanır.

image.png

 **Örnek**

 Bir yapay zeka ajanını satranç gibi bir oyun oynaması için eğittiğinizi düşünün . Temsilci farklı hareketleri araştırır ve sonuca göre olumlu veya olumsuz geri bildirim alır. Takviyeli Öğrenme aynı zamanda çevreleriyle etkileşime girerek görevleri yerine getirmeyi öğrendikleri uygulamaları da bulur.

**Takviyeli Makine Öğrenimi Türleri**

Takviyeli öğrenmenin iki ana türü vardır:

1. Pozitif Takviye
2. Negatif Takviye

**Pozitif Takviye**

* İstenilen eylemi gerçekleştiren temsilciyi ödüllendirir.

* Temsilciyi davranışı tekrarlamaya teşvik eder.

* Örnekler: Bir köpeğe oturması için ödül vermek, oyunda doğru cevap için bir puan vermek.

**Negatif Takviye**

* İstenilen bir davranışı teşvik etmek için istenmeyen bir uyaranı ortadan kaldırır.

* Temsilciyi davranışı tekrarlamaktan caydırır.

* Örnekler: Bir kola basıldığında yüksek sesli uyarının kapatılması, bir görevi tamamlayarak cezadan kaçınmak.

**Takviyeli Makine Öğreniminin Avantajları**
* Görevlere çok uygun olan ve robotik ve oyun oynama gibi bir dizi karar almayı öğrenebilen otonom karar verme mekanizmasına sahiptir.

* Uzun vadede ulaşılması çok zor olan sonuçlara ulaşmak için bu teknik tercih edilmektedir.

* Geleneksel tekniklerle çözülemeyen karmaşık problemleri çözmek için kullanılır.

** Takviyeli  Makine Öğreniminin Dezavantajları**

* Eğitim Güçlendirme Öğrenme aracıları hesaplama açısından pahalı ve zaman alıcı olabilir.

* Takviyeli öğrenme, basit problemleri çözmeye tercih edilmez.

* Çok fazla veriye ve çok fazla hesaplamaya ihtiyaç duyuyor, bu da onu kullanışsız ve maliyetli kılıyor.

Güçlendirme Makine Öğrenimi Uygulamaları
Takviyeli öğrenmenin bazı uygulamaları şunlardır:

* Oyun Oynama : RL, temsilcilere karmaşık oyunlar bile olsa oyun oynamayı öğretebilir.
* Robotik : RL, robotlara görevleri otonom olarak gerçekleştirmeyi öğretebilir.
* Otonom Araçlar : RL, sürücüsüz araçların gezinmesine ve karar vermesine yardımcı olabilir.
* Öneri Sistemleri : RL, kullanıcı tercihlerini öğrenerek öneri algoritmalarını geliştirebilir.
* Sağlık Hizmetleri : RL, tedavi planlarını ve ilaç keşfini optimize etmek için kullanılabilir.
* Doğal Dil İşleme (NLP) : RL, diyalog sistemlerinde ve sohbet robotlarında kullanılabilir.
* Finans ve Ticaret : RL algoritmik ticaret için kullanılabilir.
* Tedarik Zinciri ve Envanter Yönetimi : RL, tedarik zinciri operasyonlarını optimize etmek için kullanılabilir.
* Enerji Yönetimi : RL, enerji tüketimini optimize etmek için kullanılabilir.
* Game AI : RL, video oyunlarında daha akıllı ve uyarlanabilir NPC'ler oluşturmak için kullanılabilir.
* Uyarlanabilir Kişisel Asistanlar : RL, kişisel asistanları geliştirmek için kullanılabilir.
* Sanal Gerçeklik (VR) ve Artırılmış Gerçeklik (AR): RL, sürükleyici ve etkileşimli deneyimler oluşturmak için kullanılabilir.
* Endüstriyel Kontrol : RL, endüstriyel süreçleri optimize etmek için kullanılabilir.
* Eğitim : RL uyarlanabilir öğrenme sistemleri oluşturmak için kullanılabilir.
* Tarım : RL tarımsal operasyonları optimize etmek için kullanılabilir.
